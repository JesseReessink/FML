# Choose one of the following: k-nearest neighbor, decision tree or random forest
### i Choose for the k-nearest neighbor
###### The k-nearest neighbor is based on the idea that the class is equal to the ones that are close by. So if K = 10 it will check for the closes 9 'neighbors'.It check if the closes neighbors have the same features



```python
X = df_subset[["age", "idg","round","match","attr_o","like_o","prob_o"]] #create the X matrix
y = df_subset['dec_o'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
```

# Evaluate the predictive performance of your model on the test set


```python
from sklearn.neighbors import KNeighborsClassifier #the object class we need

knn = KNeighborsClassifier(n_neighbors=4) #create a KNN-classifier with 3 neighbors (default)
knn = knn.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
knn.score(X_test, y_test) #calculate the fit on the test data
# the score is the accurracy
```

### This shows that the accurracy of this model is 73,5%
###### now we make a confusion matrix to see how mutch the model is off


```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) #the predicted values
con = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
con
```

```python
conf_matrix = pd.DataFrame(con, index=['actual yes to a follow up date', 'actual no to follow up date'], columns = ['predicted yes', 'predicted no']) 
conf_matrix
```
